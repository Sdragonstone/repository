```python
#수집할 주소
#대표 주소
https://mypetlife.co.kr/
    
# 수집할 페이지
https://mypetlife.co.kr/popular-posts/?filter=dog # get 통신(http통신)  # post통신 (https 통신)
https://mypetlife.co.kr/popular-posts/?filter=cat    
```

```python
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
HEADERS = {"user-agent": USER_AGENT}

URL_TMPL = "https://mypetlife.co.kr/popular-posts/page/{}/?filter={}"
CATEGORY = {"dog", "cat"}
```

```python
import requests
import time
from bs4 import BeautifulSoup

import ujson
```

```python
collected_data = []

for category in CATEGORY:
    page = 1
    print("카테고리 {} 수집 시작".format(category))
    while True:
        if page % 3 == 1:
            print("현재 {}페이지 수집중...".format(page))
        req = requests.get(URL_TMPL.format(page, category), headers=HEADERS) #connection
        html = req.text # source_code
        
        soup = BeautifulSoup(html, 'lxml') #parsing (source_code texts -> soup object)
        titles = [t.get_text(strip=True) for t in soup.select("div.post-title")]
        links = [l["href"] for l in soup.select("div.post-list-title > a")]
        
        for title, link in zip(titles, links):
            collected_data.append({
                "title": title,
                "link": link,
                "category": category
            })
        if len(titles) < 10:
            break # while Loop break
        page += 1
        time.sleep(1)
    print("카테고리 {} 수집 완료".format(category))
```

```python
def soup_select(soup, css_selector, multiple=False, getText=True):
    
    res = soup.select(css_selector) if multiple else soup.select_one(css_selector)
    
    if res == [] or res is None:
        return [] if multiple else ""
    
    return [r if not getText else r.get_text(strip=True) for r in res] if multiple else res if not getText else res.get_text(strip=True)
```

```python
posts = []
for idx, data in enumerate(collected_data):
    if idx % 20 == 0:
        time.sleep(1)
        print("현재 크롤링 진행 상황 {}/{}".format(idx+1, len(collected_data)))
        
    req = requests.get(data["link"], headers=HEADERS)
    html = req.text
    
    soup = BeautifulSoup(html, 'lxml')
    
    title = soup_select(soup, "h1.post-title")
    content = soup_select(soup, "div.entry-content")
    author = soup_select(soup, "div.comment-author > b")
    date = soup_select(soup, "div.entry-header span.date")
    tags = soup_select(soup, "span.tagcloud > a", multiple=True)
    
    posts.append({
        "title" : title,
        "content" : content,
        "author" : author,
        "date" : date,
        "tags" : tags,
        "category" : data["category"]
    })
    
    time.sleep(0.5)
    
posts
```
